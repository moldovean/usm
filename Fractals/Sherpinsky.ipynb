```python
import numpy as np
import pylab
from IPython.display import Image
```

```python
#Sherpinsky Triangle
A = [[0.5, 0],[0,0.5]]
b1 = np.array([0,0])
b2 = np.array([0,10])
b3 = np.array([10,0])
```

```python
def f1(x, Points):
    Points.append(tuple(np.dot(A,x)+b1))
def f2(x, Points):
    Points.append(tuple(np.dot(A,x)+b2))
def f3(x, Points):
    Points.append(tuple(np.dot(A,x)+b3))
functions_list = [f1, f2, f3]

    
    
```

```python
# Random functions
x = (10,7)
Points= [x]
for i in range(3**10):
    np.random.choice(functions_list)(Points[-1], Points)
```

```python
pylab.figure(1)
X, Y = zip(*Points)
pylab.plot(X, Y, 'o')

pylab.savefig("Sherpinsky.png")
pylab.show()
```

```python
# The Fern Barnsley, Michael (2000). Fractals everywhere. Morgan Kaufmann.ISBN 0-12-079069-6
A1 = [[0, 0],[0,0.16]]
b1 = np.array([0,0])

A2 = [[0.85, 0.04],[-0.04, 0.85]]
b2 = np.array([0, 1.6])

A3 = [[0.2, -0.26],[0.23, 0.22]]
b3 = np.array([0,1.6])

A4 = [[-0.15, 0.28],[0.26, 0.24]]
b4 = np.array([0,0.44])


def f1(x, Points):
    Points.append(tuple(np.dot(A1,x)+b1))
def f2(x, Points):
    Points.append(tuple(np.dot(A2,x)+b2))
def f3(x, Points):
    Points.append(tuple(np.dot(A3,x)+b3))
def f4(x, Points):
    Points.append(tuple(np.dot(A4,x)+b4))
functions_list = [f1, f2, f3, f4]

```

```python
x = (0,0)
Points= [x]
for i in range(4**8):
    np.random.choice(functions_list, p=[0.01,0.85,0.07, 0.07])(Points[-1], Points)
```

```python
pylab.figure(1)
X, Y = zip(*Points)
pylab.plot(X, Y, 'o')
pylab.savefig("Fern.png")
pylab.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import animation
```

```python
fig = plt.figure()
ax = plt.axes(xlim=(0, 5), ylim=(-2, 2))
line, = ax.plot([], [], lw=2)
```

```python
def init():
    line.set_data([], [])
    return line,

def animate(i):
    x = np.linspace(0, 5, 1000)
    y = np.sin(x + 0.05*i)
    line.set_data(x, y)
    return line,
```

```python
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=620, interval=20, blit=True)
plt.show()
```
